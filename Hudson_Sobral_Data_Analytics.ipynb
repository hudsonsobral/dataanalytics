# 10. Estat√≠stica

```python
# C√°lculo da M√©dia
import numpy as np

# Exemplo com os n√∫meros: 2, 4, 6, 8, 10
numeros = [2, 4, 6, 8, 10]

# M√©todo 1: Usando fun√ß√£o built-in
media1 = sum(numeros) / len(numeros)

# M√©todo 2: Usando numpy
media2 = np.mean(numeros)

# M√©todo 3: Implementando a f√≥rmula manualmente
def calcular_media(valores):
    return sum(valores) / len(valores)

media3 = calcular_media(numeros)

print(f"N√∫meros: {numeros}")
print(f"M√©dia (m√©todo 1): {media1}")
print(f"M√©dia (m√©todo 2): {media2}")
print(f"M√©dia (m√©todo 3): {media3}")
```

```python
import numpy as np
import pandas as pd
from statistics import mode
notas = [8, 7, 6, 9, 7, 8, 10, 9, 6, 7, 8, 2, 10, 9, 7, 6, 8, 9, 5, 10]
media = np.mean(notas)
mediana = np.median(notas)
moda = mode(notas)
print("M√©dia ¬† :", media)
print("Mediana :", mediana)
print("Moda ¬† ¬†:", moda)
```

```python
amplitude = max(notas) - min(notas)
variancia = np.var(notas, ddof=1) # ddof=1 ‚Üí amostral
desvio_padrao = np.std(notas, ddof=1)
cv = desvio_padrao / media
print("Amplitude ¬† ¬† ¬† ¬†:", amplitude)
print("Vari√¢ncia ¬† ¬† ¬† ¬†:", variancia)
print("Desvio Padr√£o ¬† ¬†:", desvio_padrao)
print("Coef. Varia√ß√£o ¬† :", cv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV com separador ';'
df = pd.read_csv('/issues.txt', sep=';')

# Verificar as primeiras linhas e colunas
print("Primeiras 5 linhas do dataset:")
print(df.head())
print("\nColunas dispon√≠veis:")
print(df.columns.tolist())
print("\nShape do dataset:", df.shape)

# Contar solicita√ß√µes por departamento
departamentos_count = df['Departamento/Filial'].value_counts()
print("\nContagem por departamento:")
print(departamentos_count)

# Criar gr√°fico de pizza
plt.figure(figsize=(12, 8))
colors = plt.cm.Set3(range(len(departamentos_count)))

# Criar o gr√°fico de pizza
wedges, texts, autotexts = plt.pie(departamentos_count.values,
                                   labels=departamentos_count.index,
                                   autopct='%1.1f%%',
                                   startangle=90,
                                   colors=colors)

# Melhorar a apar√™ncia do texto
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(10)

for text in texts:
    text.set_fontsize(10)

plt.title('Distribui√ß√£o de Solicita√ß√µes por Departamento', fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')  # Para garantir que o gr√°fico seja circular

# Adicionar uma legenda
plt.legend(wedges, [f'{dept}: {count} solicita√ß√µes' for dept, count in zip(departamentos_count.index, departamentos_count.values)],
          title="Departamentos",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()

# Mostrar estat√≠sticas resumidas
print(f"\nTotal de solicita√ß√µes: {len(df)}")
print(f"N√∫mero de departamentos: {len(departamentos_count)}")
print(f"Departamento com mais solicita√ß√µes: {departamentos_count.index[0]} ({departamentos_count.iloc[0]} solicita√ß√µes)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

# Ler o arquivo CSV com separador ';'
df = pd.read_csv('/issues.txt', sep=';')

# Converter a coluna 'Data de in√≠cio' para datetime
df['Data de in√≠cio'] = pd.to_datetime(df['Data de in√≠cio'], format='%d/%m/%Y', errors='coerce')

# Remover linhas com datas inv√°lidas
df_clean = df.dropna(subset=['Data de in√≠cio'])

print(f"Total de registros com data v√°lida: {len(df_clean)}")
print(f"Per√≠odo analisado: {df_clean['Data de in√≠cio'].min().strftime('%d/%m/%Y')} a {df_clean['Data de in√≠cio'].max().strftime('%d/%m/%Y')}")

# Criar coluna com ano-m√™s para agrupamento
df_clean['Ano_Mes'] = df_clean['Data de in√≠cio'].dt.to_period('M')

# Contar solicita√ß√µes por m√™s
solicitacoes_por_mes = df_clean.groupby('Ano_Mes').size().reset_index(name='Quantidade')

# Converter per√≠odo para string para melhor visualiza√ß√£o
solicitacoes_por_mes['Mes_Str'] = solicitacoes_por_mes['Ano_Mes'].astype(str)

print("\nSolicita√ß√µes por m√™s:")
print(solicitacoes_por_mes)

# Criar o gr√°fico de linha (curva de crescimento)
plt.figure(figsize=(15, 8))

# Plotar a linha principal
plt.plot(solicitacoes_por_mes['Mes_Str'],
         solicitacoes_por_mes['Quantidade'],
         marker='o',
         linewidth=2.5,
         markersize=8,
         color='#2E86AB',
         markerfacecolor='#A23B72',
         markeredgecolor='white',
         markeredgewidth=2)

# Adicionar valores nos pontos
for i, (mes, qtd) in enumerate(zip(solicitacoes_por_mes['Mes_Str'], solicitacoes_por_mes['Quantidade'])):
    plt.annotate(f'{qtd}',
                (i, qtd),
                textcoords="offset points",
                xytext=(0,10),
                ha='center',
                fontsize=10,
                fontweight='bold',
                color='#2E86AB')

# Adicionar linha de tend√™ncia
z = np.polyfit(range(len(solicitacoes_por_mes)), solicitacoes_por_mes['Quantidade'], 1)
p = np.poly1d(z)
plt.plot(solicitacoes_por_mes['Mes_Str'], p(range(len(solicitacoes_por_mes))),
         "--", alpha=0.7, color='red', linewidth=2, label='Linha de Tend√™ncia')

# Personalizar o gr√°fico
plt.title('Evolu√ß√£o Temporal das Solicita√ß√µes de TI\n(Curva de Crescimento por M√™s)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Per√≠odo (Ano-M√™s)', fontsize=12, fontweight='bold')
plt.ylabel('N√∫mero de Solicita√ß√µes', fontsize=12, fontweight='bold')

# Rotacionar labels do eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Adicionar grid
plt.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)

# Adicionar √°rea sombreada sob a curva
plt.fill_between(solicitacoes_por_mes['Mes_Str'],
                 solicitacoes_por_mes['Quantidade'],
                 alpha=0.2, color='#2E86AB')

# Adicionar legenda
plt.legend(fontsize=10)

# Ajustar layout
plt.tight_layout()
plt.show()

# An√°lise estat√≠stica
print(f"\nEstat√≠sticas da evolu√ß√£o temporal:")
print(f"M√©dia de solicita√ß√µes por m√™s: {solicitacoes_por_mes['Quantidade'].mean():.1f}")
print(f"Mediana de solicita√ß√µes por m√™s: {solicitacoes_por_mes['Quantidade'].median():.1f}")
print(f"Desvio padr√£o: {solicitacoes_por_mes['Quantidade'].std():.1f}")

# Identificar picos e vales
max_mes = solicitacoes_por_mes.loc[solicitacoes_por_mes['Quantidade'].idxmax()]
min_mes = solicitacoes_por_mes.loc[solicitacoes_por_mes['Quantidade'].idxmin()]

print(f"\nM√™s com MAIS solicita√ß√µes: {max_mes['Mes_Str']} ({max_mes['Quantidade']} solicita√ß√µes)")
print(f"M√™s com MENOS solicita√ß√µes: {min_mes['Mes_Str']} ({min_mes['Quantidade']} solicita√ß√µes)")

# An√°lise de tend√™ncia
if z[0] > 0:
    print(f"\nTend√™ncia: CRESCENTE (coeficiente: +{z[0]:.2f} solicita√ß√µes/m√™s)")
elif z[0] < 0:
    print(f"\nTend√™ncia: DECRESCENTE (coeficiente: {z[0]:.2f} solicita√ß√µes/m√™s)")
else:
    print(f"\nTend√™ncia: EST√ÅVEL")

# Mostrar os 5 meses com mais solicita√ß√µes
top_5_meses = solicitacoes_por_mes.nlargest(5, 'Quantidade')
print(f"\nTop 5 meses com mais solicita√ß√µes:")
for idx, row in top_5_meses.iterrows():
    print(f"{row['Mes_Str']}: {row['Quantidade']} solicita√ß√µes")
```
